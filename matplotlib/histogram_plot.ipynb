# Histogram Plot

Importing necessary libraries

```python
import numpy as np
from matplotlib import pyplot as plt
import random
```

```python
x =  np.random.randint(10,60,(50))
print(x)
```

```python
plt.hist(x)
plt.title('Histogram')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

Colouring

```python
plt.hist(x, color='m', edgecolor='k')
plt.title('Histogram')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

Categoraizing

```python
l = range(10,70,10)
plt.hist(x, color='y', edgecolor='k', bins=l)
plt.title('Histogram')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
l = range(10,70,10)
plt.hist(x, color='m', edgecolor='k', bins='auto', range=(0,100))
plt.title('Histogram')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
l = range(10,70,10)
plt.hist(x, color='m', edgecolor='k', bins=l, cumulative=-1)
plt.title('Histogram')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
