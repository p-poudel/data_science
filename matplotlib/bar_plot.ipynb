# Bar Plot

#### Importing pyplot from Matplotlib

We can also use 'from matplotlib import pyplot as plt'

```python
import matplotlib.pyplot as plt
```

```python
x = ['python', 'c', 'c++', 'java']
y = [85,70,65,83]
```

We use plt.bar() to plot par graph

plt.show() is used to show generated plot(diagram).

```python
plt.bar(x,y)
plt.show()
```

Labelling the axis

```python
plt.xlabel('Language', fontsize=15)
plt.ylabel('Percentage of Users',fontsize=15)
plt.title('Barplot',fontsize=24)
plt.bar(x,y)
plt.show()
```

Changing the width and colour of the bar graph

Changes is done in plt.bar()

```python
plt.xlabel('Language', fontsize=15)
plt.ylabel('Percentage of Users',fontsize=15)
plt.title('Barplot',fontsize=24)
plt.bar(x,y, width=0.4, color=['c','m','y','k'])
plt.show()
```

We can change the alignment of text on x-asis by align attribute of plt.bar()

it's value can be 'edge', 'center'(default)

```python
plt.xlabel('Language', fontsize=15)
plt.ylabel('Percentage of Users',fontsize=15)
plt.title('Barplot',fontsize=24)
plt.bar(x,y, width=0.4, color=['c','m','y','k'], align='edge')
plt.show()
```

Changing the boundary of the bars:

We can change their colours, boundary colours, boundary styles and width using:

0<alpha<1

```python
plt.xlabel('Language', fontsize=15)
plt.ylabel('Percentage of Users',fontsize=15)
plt.title('Barplot',fontsize=24)
plt.bar(x,y, width=0.4, color=['c','m','y','k'], edgecolor='r', linewidth=3, linestyle=':', alpha=0.4)
plt.show()
```

Using Labels and Legends

Legends is required to show Labels

```python
z = [20,30,50,45]
plt.xlabel('Language', fontsize=15)
plt.ylabel('Percentage of Users',fontsize=15)
plt.title('Barplot',fontsize=24)
plt.bar(x,y, width=0.4, label='Popularity')
plt.bar(x,z, width=0.4, label='Previous Popularity')
plt.legend()
plt.show()
```

Separating above graphs

```python
import numpy as np

p=np.arange(len(x))
width = 0.2
p1 = [j+width for j in p]

plt.xlabel('Language', fontsize=15)
plt.ylabel('Percentage of Users',fontsize=15)
plt.title('Barplot',fontsize=24)

plt.bar(p,y, width, label='Popularity')
plt.bar(p1,z, width, label='Previous Popularity')

plt.xticks(p+(width/2),x,rotation=90)
plt.legend()
plt.show()
```

## Horizontal Bar Graph

```python
plt.barh(p,y, width, label='Popularity')
plt.barh(p1,z, width, label='Previous Popularity')

plt.yticks(p+(width/2),x,rotation=90)
plt.legend()
plt.show()
```
