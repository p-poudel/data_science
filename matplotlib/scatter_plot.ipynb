# Scatter Plot

Importing basic libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,8)
y = [i**3 for i in x]
```

```python
plt.scatter(x,y)
plt.title('Scatter Plot', fontsize=24)
plt.xlabel('Number')
plt.ylabel('Cube')
plt.show()
```

Changing Colours

```python
plt.scatter(x,y,color='r')
plt.title('Scatter Plot', fontsize=24)
plt.xlabel('Number')
plt.ylabel('Cube')
plt.show()
```

```python
plt.scatter(x,y,c=['c','m','y','k','r','g','b']) # can use color or c
plt.title('Scatter Plot', fontsize=24)
plt.xlabel('Number')
plt.ylabel('Cube')
plt.show()
```

Changing the size of the points of scatter

```python
color = ['c','m','y','k','r','g','b']
plt.scatter(x,y,c=color, s=[10*i for i in y])
plt.title('Scatter Plot', fontsize=24)
plt.xlabel('Number')
plt.ylabel('Cube')
plt.show()
```

Changing Texture/Transparency

Same like in bar_plot we can use alpha

```python
color = ['c','m','y','k','r','g','b']
plt.scatter(x,y,c=color, s=[10*i for i in y], alpha=0.3)
plt.title('Scatter Plot', fontsize=24)
plt.xlabel('Number')
plt.ylabel('Cube')
plt.show()
```

Changing the marker from point to other objects

```python
color = ['c','m','y','k','r','g','b']
plt.scatter(x,y,c=color, s=[10*i for i in y],marker='*', edgecolors='grey', linewidths=2)
plt.title('Scatter Plot', fontsize=24)
plt.xlabel('Number')
plt.ylabel('Cube')
plt.show()
```

We can use Color Codes (0<100) instead of colours using cmap

```python
color = np.random.default_rng().integers(low=1,high=100,size=7)
plt.scatter(x,y,c=color, s=[10*i for i in y],cmap="BrBG")
plt.colorbar().set_label('ColorBar')
plt.title('Scatter Plot', fontsize=24)
plt.xlabel('Number')
plt.ylabel('Cube')
plt.show()
```

Multiple Plots

```python
z = [(8-i)**2 for i in x]

plt.scatter(x,y,color='g', s=[10*i for i in y])
plt.scatter(x,z,color='r')
# plt.colorbar().set_label('ColorBar')

plt.title('Scatter Plot', fontsize=24)
plt.xlabel('Number')
plt.ylabel('Cube')
plt.show()
```
